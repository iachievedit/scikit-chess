# Load `chess.dat` data from disk

```python
import pickle
file  = open("chess.dat", "rb")
chess = pickle.load(file)
X,y = chess["data"], chess["target"]
print("Loaded %d images" % len(X))
```

# Shuffle the data

```python
from sklearn.utils import shuffle
X, y = shuffle(X, y, random_state=42)
```

# Pick a random piece

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import random

i = random.randrange(0,len(X))
some_piece = X[i]
some_piece_image = some_piece.reshape(72,72)

some_piece_classification = y[i]

plt.imshow(some_piece_image)
plt.axis("off")
plt.show()

print("The image above should be a %s " % (some_piece_classification))
```

# Specify training set

```python
# Of our 6912 images, the first 5500 will be our training set
X_train, X_test, y_train, y_test = X[:5500], X[5500:], y[:5500], y[5500:]

import numpy as np
shuffle_index = np.random.permutation(5500)
X_train, y_train = X_train[shuffle_index], y_train[shuffle_index]
```

# Train!

```python
# A One versus All classifier
from sklearn.linear_model import SGDClassifier
sgd_clf = SGDClassifier(random_state=42, max_iter=1000, tol=1e-3)
sgd_clf.fit(X_train, y_train)
```

# How did we do?

```python
import random
from chesspieces import ChessPieces

for i in range(0,24):
    j = random.randrange(5500,len(X))
    some_piece = X[j]
    some_piece_image = some_piece.reshape(72,72)
    print(some_piece_image.shape)

    plt.imshow(some_piece_image)
    plt.axis("off")
    plt.show()

    print("This is a %s." % ChessPieces[sgd_clf.predict([some_piece])[0]])
```
