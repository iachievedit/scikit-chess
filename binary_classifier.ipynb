* Load the `chess.dat` data from disk

```python
import pickle
file  = open("chess.dat", "rb")
chess = pickle.load(file)
X,y = chess["data"], chess["target"]
```

* Shuffle

```python
from sklearn.utils import shuffle
X, y = shuffle(X, y, random_state=42)
```

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

some_piece = X[3000]
some_piece_image = some_piece.reshape(72,72)

some_piece_classification = y[3000]

plt.imshow(some_piece_image)
plt.axis("off")
plt.show()

print("The image above should be a %s " % (some_piece_classification))
```

```python
# Of our 6912 images, the first 5500 will be our training set
X_train, X_test, y_train, y_test = X[:5500], X[5500:], y[:5500], y[5500:]
```

```python
import numpy as np
shuffle_index = np.random.permutation(5500)
trainingPiece = 'BB'
X_train, y_train = X_train[shuffle_index], y_train[shuffle_index]
y_train_piece = (y_train == trainingPiece)
y_test_piece  = (y_test  == trainingPiece)
```

```python
# A binary classifier that returns true for the trainingPiece
from sklearn.linear_model import SGDClassifier
sgd_clf = SGDClassifier(random_state=42, max_iter=1000, tol=1e-3)
sgd_clf.fit(X_train, y_train_piece)
```

```python
for i in [5517, 5520, 5521, 5527, 5533, 5539, 5546, 5547, 5549, 5555, 5560, 5572]:
    some_piece = X[i]
    some_piece_image = some_piece.reshape(72,72)

    plt.imshow(some_piece_image)
    plt.axis("off")
    plt.show()

    # Returns True if the image is thought to be the trainingPiece, false otherwise
    print("Is this a %s?" % trainingPiece)
    if sgd_clf.predict([some_piece]):
        print("Yes!")
    else:
        print("Probably not")
```
